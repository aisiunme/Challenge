#### 자주 쓰이는 항목명 삭제 함수 만들기

```python
def pre_processing(work_text) :
    return work_text.replace("\n", "").replace("\t", "").replace("\r","").replace("  (", "(").replace("\xa0","")
```

#### 기업 공시 불러와 리스트 만드는 함수 생성 

```python
def make_anounce_list() :
    
    # 사이트 요청하여 불러오기
    
    url = "http://dart.fss.or.kr/dsac001/mainAll.do"
    res = requests.get(url)
    soup = bs(res.text, 'html.parser')

    # tr리스트 만들어 반복문 생성
    
    tr_list = soup.findAll('tr')
    
    
    announce_list = []
     
    
    
    for item in tr_list[1:50] :    # tr_list 의 첫번째 항목은 테이블의 항목명임. 그래서 2번째 항목부터 작업을 진행해야 함. 10번째까지 진행
        
        work1 = item.findAll('td')
        
        # 각 항목별로 데이터 추출
        
        company = pre_processing(work1[1].text)
        market = work1[1].img['title']
        announce_date = work1[4].text
        announce_time = pre_processing(work1[0].text)
        title = pre_processing(work1[2].text)
        sub_url = work1[2].a['href']         
        
        # 코넥스, 기타법인 제외 and 기재정정 공시는 제외
        
        if (market not in ['코넥스시장', '기타법인']) and ('기재정정' not in title):
        
            # 코드 구하기
            code_url = work1[1].a['href']
            code_request = requests.get('http://dart.fss.or.kr'+code_url)
            code_soup = bs(code_request.text, 'html.parser')
            code = pd.read_html(code_request.text)[0].iloc[3,1]
        
            # announce_list 에 정리하기

            announce_list.append({
                'company' : company,
                'market' : market,
                'announce_date' : announce_date, 
                'announce_time' : announce_time,
                'title' : title,
                'announce_url' : 'http://dart.fss.or.kr' + sub_url, 
                'info_url' : "http://finance.naver.com/item/main.nhn?code="+ code,
                'chart' : 'https://ssl.pstatic.net/imgfinance/chart/item/candle/day/{0}.png'.format(code)
            })

    return announce_list
```

#### main 함수 정의

- announce 함수를 통해 announce list 생성(코넥스, 기타법인이 아닌 리스트)


- 발송 리스트(sent_list)에 없을 경우 텔레그램 발송


- 발송한 내역을 발송 리스트에 append함



```python
def main() :
    announce_list = make_anounce_list()   
    for item in announce_list :
        
        # 각 공시 item의 title이 condition_list에 있는지 체크
        for condition in condition_list :
            
            # title 내용이 condition_list 에 포함된다면, 그리고 item이 이미 sent_list에 없다면,
            if (condition in item['title']) and (item not in sent_list) : 
            #텔레그램 발송
                send_message(item['announce_date'], item['announce_time'], item['company'], item['title'], item['announce_url'], item['info_url'], item['chart'])
                sent_list.append(item)
                pass   # 반드시 pass를 붙여야 함. 그렇지 않으면 다른 키워드 때 중복 발생
```
